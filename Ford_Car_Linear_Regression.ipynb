```python
import pandas as pd
# Load dataset
df = pd.read_csv("ford.csv")

# Independent Features
X = df.drop("price", axis=1)

# Dependent Feature
y = df["price"]

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

Observation:

The dataset was loaded successfully.

price was selected as the target variable (y).

Remaining columns were selected as independent features (X).

```python
# Identify categorical columns
cat_cols = X.select_dtypes(include="object").columns

print("Categorical Columns:")
print(cat_cols)

# One-Hot Encoding
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

# Convert Boolean columns to Integer
X = X.astype(int)

# First 5 rows
print(X.head())
```

Observation:

The categorical columns identified were model, transmission, and fuelType.

One-Hot Encoding converted these text-based columns into numerical (0 and 1) columns.

The encoded features were converted to integer type using astype(int).

The dataset is now in a numerical format and ready for machine learning algorithms.

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

scaler = StandardScaler()

X[num_cols] = scaler.fit_transform(X[num_cols])

print(X.head())
```

Observation:

Standard Scaling was applied to numerical features.

The scaled values now have approximately mean = 0 and standard deviation = 1.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42
)

print("X_train Shape:", X_train.shape)
print("X_test Shape :", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape :", y_test.shape)
```

Observation:

The dataset was split into training (67%) and testing (33%) sets.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print("Intercept:")
print(model.intercept_)

print("\nCoefficients:")
print(model.coef_)
```

Observation:

The Linear Regression model was trained successfully.

The coefficients show how each feature influences the car price.

```python
# Predictions
y_pred = model.predict(X_test)

comparison = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": y_pred
})

print(comparison.head(10))
```

Observation:

The predicted prices are close to the actual prices.

Small differences are expected due to prediction error.

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

Interpretation

The R² score measures how well the model explains the variation in car prices.

A value closer to 1 indicates better performance.

Observation:

A higher R² score means the model predicts car prices accurately.

If the score is close to 1, the model performs well.

```python
import joblib

# Save Model
joblib.dump(model, "LR_ford_car.pkl")

# Save Scaler
joblib.dump(scaler, "scaler.pkl")

# Save Feature Columns
joblib.dump(X.columns.tolist(), "columns.pkl")

print("Model and preprocessing objects saved successfully.")
```

Observation:

The trained Linear Regression model was saved as LR_ford_car.pkl.

The StandardScaler was saved as scaler.pkl.

The feature column names were saved as columns.pkl for future predictions.
